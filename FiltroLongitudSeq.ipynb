```python
# Importación de Modulos
import Bio
from Bio import SeqIO, SearchIO, Entrez
from Bio.Seq import Seq
import fastaparser

print("Biopython version:", Bio.__version__)
```

```python

FullSearch = 'BusquedaCompleta.fasta'

for record in SeqIO.parse(FullSearch, "fasta"):
  print(record.id, '    length: ', len(record.seq), '\n ',repr(record.seq))
```

```python
print('El archivo fasta de la busqueda de proteínas contiene')
!grep -c '>' BusquedaCompleta.fasta
print('secuencias') 
```

```python
# Selección de archivos fasta que poseen 50 o menos aminoácidos en su secuencia
for record in SeqIO.parse(FullSearch, "fasta"):
  if len(record.seq) <= 50:
    with open('FilterFasta.fasta', 'a') as handle:
      SeqIO.write(record, handle, 'fasta')
```

```python
# Descripción de los archivos seleccionados 
for record in SeqIO.parse('FilterFasta.fasta', "fasta"):
  print(record.id, '    length: ', len(record.seq))
  print('  ',repr(record.seq))
```

```python
print('El archivo fasta con el filtro por longitud de aminoácidos contiene')
!grep -c '>' FilterFasta.fasta
print('secuencias') 
```
